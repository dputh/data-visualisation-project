```python
import pandas as pd
import seaborn as sns
import datetime as dt

file = open("./data/watch-history.json")
data = pd.read_json(file)
data = data.drop("details", axis=1)
data = data.drop("description", axis=1)

data = data[  data["titleUrl"].notna()  ]
data = data[  data["subtitles"].notna()  ]

data
```

```python
import re

data["date_time"] = pd.to_datetime(data["time"])
data["time"] = data["date_time"].map(lambda x: x.time())
data["day"] = data["date_time"].map(lambda x: x.day)
data["week"] = data["date_time"].map(lambda x: x.week)
data["month"] = data["date_time"].map(lambda x: x.month)
data["year"] = data["date_time"].map(lambda x: x.year)
data["channel"] = data["subtitles"].map(lambda x: x[0]["name"])
data["titleUrl"] = data["titleUrl"].map(     lambda x: (re.search("(?<=v=).+", x).group(0))    )


data.head(5)
```

```python
import numpy as np
import time
import requests
import math
import pandas as pd


requested_data = pd.DataFrame()


requested_data["titleUrl"]=data["titleUrl"]
requested_data["tags"] = -1

N = len(requested_data["titleUrl"].tolist())
print(N)
for k in np.arange(1504,N):


    vid_id = requested_data["titleUrl"][k]
    key = "AIzaSyBZpYJo1SjGE_eGTze3RclPnDQoHony9hg"

    URL = 'https://www.googleapis.com/youtube/v3/videos?key={}&fields=items(snippet(title,description,tags))&part=snippet&id={}'.format(key, vid_id)
    r = requests.get(URL)
    
    json = r.json()
    try:
        tags = json["items"][0]["snippet"].get("tags", 0)
    except:
        tags = 0

    requested_data["tags"][k] = tags
    requested_data.to_csv("requested_data.csv")

    print("Iteration {}, tags added {}".format(k, tags))

    time.sleep(5)
```
