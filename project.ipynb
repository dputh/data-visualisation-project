```python
import pandas as pd
import seaborn as sns
import datetime as dt

file = open("./data/watch-history.json")
data = pd.read_json(file)
data
```

```python
data.drop("details", "description")
data
```

```python
data["date_time"] = pd.to_datetime(data["time"])
data["time"] = data["date_time"].map(lambda x: x.time())
data["day"] = data["date_time"].map(lambda x: x.day)
data["month"] = data["date_time"].map(lambda x: x.month)
data["year"] = data["date_time"].map(lambda x: x.year)
data["channel"] = data["subtitles"][0:5].map(lambda x: x[0]["name"])
```
